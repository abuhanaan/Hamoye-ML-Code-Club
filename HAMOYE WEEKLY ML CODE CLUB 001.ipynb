CELL PHONE IS MATCHED WITH POPULATION SUCH THAT WE HAVE 11640 ROWS AND 4 COLUMNS

```python
joined_df = cell_phones.merge(population, how='left', on=('Country', 'year'))


joined_df.head()
```

```python
joined_df.info()
```

RENAMING COLUMNS IN THE GEO_CODE DATAFRAME

```python
geo_code.rename(columns={'country':'geo', 'name':'Country'}, inplace=True)

geo_code.head()
```

```python
# Creating a new df that comprises the geo and the Country columns from the geo_code dataframe
new_geo = geo_code[['geo', 'Country']]

new_geo.head()
```

```python
new_geo.info()
```

```python
# merging the joined_df with new_geo
final_joined_df = joined_df.merge(new_geo, how='inner', on='Country')

final_joined_df.head()
```

```python
final_joined_df.info()
```

```python
# Creating a new feature ---> cell_phones_per_person

final_joined_df['cell_phones_per_person'] = final_joined_df['cell_phones_total']/final_joined_df['population']
```

```python
final_joined_df.head()
```

```python
# Converting the geo column to upper case

final_joined_df['geo'] = final_joined_df['geo'].str.upper()

final_joined_df.head()
```

```python
final_joined_df.describe()
```

```python
# Asserting that the dataset comprises of 194 countries

assert(len(final_joined_df['Country'].unique()) == 194)
```

```python
# Saving countries in 2017 to a variable
countries_in_2017 = final_joined_df[['Country', 'cell_phones_total']].loc[final_joined_df['year']==2017]

countries_in_2017.head()
```

```python
# Getting the top 5 countries in 2017 in terms of cell_phones total
countries_in_2017 = countries_in_2017.sort_values(by=['cell_phones_total'], ascending=False)
top_5_countries_in_2017 = countries_in_2017.head()

top_5_countries_in_2017
```
